Use `curve_fit` in `scipy` library to fit the three coefficients, $c_0$, $c_1$, and $c_2$. Plot the obtained curve with the scatter points.

````python
### Put your code here
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Extract variables
x = df['Perimeter'].values
y = df['ConvexArea'].values

# Define quadratic model
def quadratic(x, c0, c1, c2):
    return c0 + c1 * x + c2 * x**2

# Fit curve
params, _ = curve_fit(quadratic, x, y)
c0, c1, c2 = params

# Printing coefficients
print(f"Fitted coefficients: c0 = {c0:.3f}, c1 = {c1:.3f}, c2 = {c2:.6f}")

# Generating the x values for plotting the curve
x_fit = np.linspace(min(x), max(x), 300)
y_fit = quadratic(x_fit, *params)

# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.5, label='Data',color = 'red')
plt.plot(x_fit, y_fit, color='blue', linewidth=2, label='Quadratic Fit')
plt.xlabel('Perimeter')
plt.ylabel('ConvexArea')
plt.title('Quadratic Fit: ConvexArea vs Perimeter')
plt.legend()
plt.grid(True)
plt.show()
````

````python
import random 
import numpy as np
import math
import matplotlib.pyplot as plt
import time  
from IPython.display import display, clear_output
````

````python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from matplotlib.cbook import get_sample_data

def imscatter(x, y, image, ax=None, zoom=1):
    if ax is None:
        ax = plt.gca()
    try:
        image = plt.imread(image)
    except TypeError:
        # Likely already an array...
        pass
    im = OffsetImage(image, zoom=zoom)
    x, y = np.atleast_1d(x, y)
    artists = []
    for x0, y0 in zip(x, y):
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', \
                            frameon=False)
        artists.append(ax.add_artist(ab))
    ax.update_datalim(np.column_stack([x, y]))
    ax.autoscale()
    return artists
````

**Human class**

````python
class human():
    
    ## initiate and parameters
    def __init__(self, roam_range, initial_x, initial_y, \
                 lower_bond, upper_bond, image_path):
        self.range = roam_range
        self.lw = lower_bond
        self.up = upper_bond
        self.image_path = image_path
        self.x = initial_x
        self.y = initial_y    

    ## draw images  
    def draw(self):
        imscatter(self.x, self.y, self.image_path, zoom=0.3)  
        
    ## random roaming  
    def roam(self):  
        self.dx = random.randint(-self.range, self.range)
        self.dy = random.randint(-self.range, self.range)
    
        # Clamp x and reverse direction if out of bounds
        if self.x <= self.lw:
            self.x = self.lw
            self.dx = abs(self.dx)
        elif self.x >= self.up:
            self.x = self.up
            self.dx = -abs(self.dx)
    
        # Clamp y and reverse direction if out of bounds
        if self.y <= self.lw:
            self.y = self.lw
            self.dy = abs(self.dy)
        elif self.y >= self.up:
            self.y = self.up
            self.dy = -abs(self.dy)
    
        self.x += self.dx
        self.y += self.dy        
````

**Zombie class** 

&#9989; **Question 5.2 (2 point)**:
Read through the code and explain what part I, part II, part III, and part IV are doing.

&#9989; **Do this**:

````python
class zombie():
    
    ## initiate and parameters
    def __init__(self, roam_range, hunt_radius, kill_rate, \
                 initial_x, initial_y, lower_bond, upper_bond, image_path):
        self.range = roam_range
        self.radius = hunt_radius
        self.kr = kill_rate
        self.x = initial_x
        self.y = initial_y
        self.lw = lower_bond
        self.up = upper_bond
        self.image_path = image_path

    ## draw images    
    def draw(self):
        imscatter(self.x, self.y, self.image_path, zoom=0.3)  
        
    ## random roaming   
    def roam(self):
        self.dx = random.randint(-self.range, self.range)
        self.dy = random.randint(-self.range, self.range)
    
        # Clamp x and reverse direction if out of bounds
        if self.x <= self.lw:
            self.x = self.lw
            self.dx = abs(self.dx)
        elif self.x >= self.up:
            self.x = self.up
            self.dx = -abs(self.dx)
    
        # Clamp y and reverse direction if out of bounds
        if self.y <= self.lw:
            self.y = self.lw
            self.dy = abs(self.dy)
        elif self.y >= self.up:
            self.y = self.up
            self.dy = -abs(self.dy)
    
        self.x += self.dx
        self.y += self.dy

    ## zombie hunts human
    def hunt(self, human):
        ## part I
        self.h_x = human.x
        self.h_y = human.y

        ## Part II
        dst = math.sqrt((self.x-self.h_x)**2 + (self.y-self.h_y)**2)
        
        catch = False 
        ## Part III
        if dst < self.radius:
            if random.random() < self.kr:
                catch = True
        return catch
        
    ## Part IV    
    def infect(self):
        new_zombie = zombie(self.range, self.radius, self.kr, \
                            self.x, self.y, self.lw, self.up, \
                            self.image_path) 
        return new_zombie
      
````
