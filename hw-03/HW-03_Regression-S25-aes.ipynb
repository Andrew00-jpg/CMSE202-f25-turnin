# Part 3: One Variable Linear and Polynomial Regression (28 points)

In this part, we'll perform some one-variable linear and polynomial regression analysis on the California Cooperative Oceanic Fisheries Investigations oceanographic and larval fish data.

&#9989;  **Question 3.1 (6 points)**: Using the `OLS` method in `statsmodels`, perform a linear regression to predict the `Salnty` using the `T_degC` and **display the results summaries**. Remember that you may need to use the `add_constant()` method to make sure `OLS` fits a general line $y = ax+b$ to the data instead of a line through the origin $y = ax$. 

````python
# Put your code here
import statsmodels.api as sm

# Define the features 
X = oceanographic_data['T_degC']
y = oceanographic_data['Salnty']

# Add a constant to the feature for the intercept term
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X)
results = model.fit()

# Display the summary
print(results.summary())
````

&#9989;  **Question 3.2 (4 points)**: Answer the following questions: 

1. What is the R-squared value you got? 

2. Based on your R-squared value, what does it tell you about the regression fit, and how the model fits the data?

<font size=+3>&#9998;</font> Put your answers here The R-squared value is 0.970.An R-squared value of 0.970 means that 97% of the variance in salinity (Salnty) can be explained by the water temperature (T_degC) using this linear regression model. This is a very strong fit, suggesting that temperature is a highly predictive factor for salinity in this dataset.

&#9989;  **Question 3.3 (6 points)**: Now make a scatter plot of `T_degC` (x-axis) vs. `Salnty` (on y-axis). Plot the best fit line on the same plot. Label the axes, and add a legend, and give the plot a title.

````python
# Put your code here
# Fit the model
import matplotlib.pyplot as plt

x = oceanographic_data["T_degC"]
y = oceanographic_data["Salnty"]

x_with_const = sm.add_constant(x)
model = sm.OLS(y, x_with_const).fit()

# Predict values for the best fit line
x_sorted = x.sort_values()
x_sorted_with_const = sm.add_constant(x_sorted)
y_pred = model.predict(x_sorted_with_const)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(x, y, color='blue', label='Observed Data')
plt.plot(x_sorted, y_pred, color='red', label='Best Fit Line')
plt.xlabel('Temperature (C)')
plt.ylabel('Salinity (PSU)')
plt.title('Linear Regression: Temperature vs. Salinity')
plt.legend()
plt.grid(True)
plt.show()
````

&#9989;  **Question 3.4 (2 points)**: What is the slope and intercept of your fit line?

````python
# Put code here.
intercept = model.params[0]
slope = model.params[1]

print("Intercept:", intercept)
print("Slope:", slope)
````

**&#9989; Do this:** **Question 3.5 (4 points)**: Use `plot_regress_exog` to investigate the distribution of residuals in your model fit. Make sure to create a large enough figure so that everything is easily visible. 

````python
# Put code here.
fig = plt.figure(figsize=(12, 8))  # Adjust the size to make it clearly visible
sm.graphics.plot_regress_exog(model, 'T_degC', fig=fig)
plt.tight_layout()
plt.show()
````

````python
# Put your code here
data = pd.read_csv("Student_Performance.csv")

data['Extracurricular Activities'] = (
    data['Extracurricular Activities']
    .astype(str)  # Convert to string in case there are NaNs
    .str.strip()  # Remove leading/trailing spaces
    .str.upper()  # Standardize case
    .map({'YES': 1, 'NO': 0})  # Map YES to 1, NO to 0
)

print(data['Extracurricular Activities'].unique())
print(data.dtypes)
````

&#9989;  **Question 4.2 (5 points)**: Using the `OLS` method in `statsmodels`, perform a multivariable linear regression to predict the `Performance Index` based on `Hours Studied`, `Previous Scores`,`Extracurricular Activities`, `Sample Question Papers Practiced` and `Sleep Hours`.
````python
# Put your code here
# Choose three independent variables
X = data[['Hours Studied', 'Previous Scores', 'Sleep Hours']]

# Add a constant to include an intercept in the model
X = sm.add_constant(X)

# Define the dependent variable
y = data['Performance Index']

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Display the summary of regression results
print(model.summary())
````

&#9989;  **Question 4.6 (3 points)**: Create five `.graphics.plot_regress_exog` figures, one for each of the features (columns of original dataframe) in your model. Pay attention to the top two plots: the fitted values figure and the residual plot. 

````python
# Put your code here.
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.graphics.api as smg

# Set up the independent and dependent variables
X = data.drop('Performance Index', axis=1)
X = sm.add_constant(X)
y = data['Performance Index']

# Fit the model
model = sm.OLS(y, X).fit()

# Features to visualize
features = ['Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced']

# Generate and display the regression plots for each feature
for feature in features:
    fig = plt.figure(figsize=(10, 6))
    smg.plot_regress_exog(model, feature, fig=fig)
    plt.tight_layout()
    plt.show()
````
